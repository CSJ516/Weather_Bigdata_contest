## 2.3 10만건당 건수 예측 모델링 (AutoML)

대분류명, 평균습도, 구매건수 변수를 제외하고 계절별로 간단하게 모델링 진행

```python
# 대분류명, 평균습도, 구매건수 제외 및 결측치 제거
beauty2 = beauty.drop(['대분류명', '평균습도(%rh)', '구매건수'], axis=1)
beauty2.dropna(inplace=True)
beauty2.reset_index(drop=True, inplace=True)
print(beauty2.shape)
beauty2.head(3)
```

```python
봄 = beauty2[beauty2['계절']=='봄'].reset_index(drop=True)
print(봄.shape)
봄.head(3)
```

```python
# 봄 데이터 이상치 확인
def outliers(data):
    q1, q3 = np.percentile(data, [25,75])
    iqr = q3 - q1
    upper_bound = q3 + (iqr * 1.5)
    lower_bound = q1 - (iqr * 1.5)
    return np.where((data > upper_bound) | (data < lower_bound))
```

```python
강수량_idx = outliers(봄['평균일강수량(mm)'])[0]
미세먼지_idx = outliers(봄['일 미세먼지 농도(㎍/㎥)'])[0]
풍속_idx = outliers(봄['평균풍속(km/h)'])[0]
```

```python
# 이상치 평균으로 처리
강수량_mean = 봄['평균일강수량(mm)'].mean()
봄.loc[강수량_idx, '평균일강수량(mm)'] = 강수량_mean

미세먼지_mean = 봄['일 미세먼지 농도(㎍/㎥)'].mean()
봄.loc[미세먼지_idx, '일 미세먼지 농도(㎍/㎥)'] = 미세먼지_mean

풍속_mean = 봄['평균풍속(km/h)'].mean()
봄.loc[풍속_idx, '평균풍속(km/h)'] = 풍속_mean

print(봄.shape)
봄.head(3)
```

```python
# Target 값 분포 확인 및 로그 변환
sns.distplot(봄['10만건당 건수'])
```

```python
봄['10만건당 건수'] = np.log1p(봄["10만건당 건수"])
sns.distplot(봄['10만건당 건수'])
```

```python
from pycaret.regression import *

setup(data=봄, target='10만건당 건수', train_size=0.8, categorical_features='일')
```

```python
reg = compare_models(fold=3, sort='RMSLE', n_select=1)
```

```python
plt.rcParams['font.family'] = 'Malgun Gothic'
plot_model(estimator=reg, plot='feature')
```
