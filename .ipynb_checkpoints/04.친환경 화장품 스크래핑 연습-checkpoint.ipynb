```python
# 최종 스크래핑 코드
def cosmetic_scraping(query):
    # webdriver 경로
    chromedriver = 'C:\\Users\\dalgo\\OneDrive\\바탕 화면\\날씨 빅데이터 공모전\\chromedriver.exe'

    # selenium의 webdriver에 연동
    driver = webdriver.Chrome(chromedriver)
    driver.get(f'https://m.lalavla.com/service/search/searchResult.html?searchKeyword={query}')

    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')
    
    # 검색 건수
    count = int(soup.find('span', class_='point01').text)
    
    data = []
    for i in range(count):
        # 브랜드명
        brand = soup.find_all('span', class_='category')[i].text
        
        # 상품명
        item = soup.find_all('span', class_='prd-name')[i].text
        
        # 가격 (정가, 할인가)
        price = soup.find_all('span', class_='price')[i].text.split('원')
        if len(price) == 3:
            pre_price, dis_price = price[0], price[1]
        else:
            pre_price = price[0]

        # 이미지
        image = driver.find_elements_by_tag_name('img')[i+1]
        image.screenshot('./image/' + f'랄라블라{str(i+1)}' + '.png')
        
        data.append([brand, item, pre_price, dis_price])
    
    df = pd.DataFrame(data)
    df.columns = ['브랜드명', '상품명', '정가', '할인가']
    return df
```

```python
cosmetic_scraping('비건')
```
